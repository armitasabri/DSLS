```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
```

```python
X = pd.DataFrame(data.data, columns=data.feature_names)
y = pd.Series(data.target)
```
