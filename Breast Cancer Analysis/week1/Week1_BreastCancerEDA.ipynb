```python
#import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
```

```python
# load the data
data = load_breast_cancer()
```

```python
# separate features from label(target)
X = pd.DataFrame(data.data, columns=data.feature_names)
y = pd.Series(data.target)
```

```python
# histogram of a numeric feature

plt.hist(X.iloc[:,0],bins=30)
plt.title('distribuation of '+ X.columns[0])
plt.show()
```

```python
plt.hist(X.iloc[:,1],bins=30)
plt.title('distribuation of '+X.columns[1])
plt.show()
```

```python
plt.figure(figsize=(20, 50))
for i in range(X.shape[1]):
    plt.subplot(15,3,i+1)
    plt.hist(X.iloc[:,i],bins=30)
    plt.title('distribuation of '+X.columns[i])
plt.tight_layout()
plt.show()
```

```python
#correlation heatmap

sns.heatmap(X.corr(),cmap='coolwarm')
plt.show()
```
